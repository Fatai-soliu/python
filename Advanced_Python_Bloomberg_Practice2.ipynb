```python
"""
Exercise 2: Grouping and Aggregating Data
"""
import pandas as pd

data = {
    'store_id': [1, 2, 1, 3, 2, 3, 1, 3, 2],
    'product': ['A', 'B', 'A', 'C', 'B', 'A', 'C', 'A', 'C'],
    'quantity_sold': [10, 20, 15, 5, 25, 30, None, 15, None],
    'unit_price': [100, 200, 100, 150, 200, 100, 150, 100, 150]
}

data

#Create dataframe
df7 = pd.DataFrame(data)

#Fill in missing values for quantity solde with the average quantity sold for that specific product

df7['quantity_sold'] = df7.groupby('product')['quantity_sold'].transform(lambda x: x.fillna(x.mean()))

#Add a new column total sales

df7['total_sales'] = df7['quantity_sold'] * df7['unit_price']

#Group by store_id and product and also sort

group_df7 = df7.groupby(['store_id','product'])[['quantity_sold','total_sales']].sum().sort_values(by='total_sales', ascending=False)
group_df7
```

## Array Manipulation

```python
##Example 1

"""
Given an array [1, 2, 3, 4, 5], rotate the array by 2 positions to the right:
"""
## The appraoch will, involve slicing the array

arr = [1, 2, 3, 4, 5]

def rotate_array(arr, k):
  # calculate the effective number of rotations
  n = len(arr)
  k = k % n # ensure k is within bounds of the array length

  #Rotate the array by slicing
  rotated_array = arr[-k:] + arr[:-k]  # The last element (arr[-k:]) + the first n - k element (arr[:-k])

  return rotated_array
```

```python
# Apply the function

rotated_arr = rotate_array(arr, 2)
print(rotated_arr)
```

Slicing is the most effective way of handling array rotations in python

```python
## In-place method - uses O(1) space complexity (constant space complexity)

def reverse(arr, start, end):
  while start < end:
    arr[start], arr[end] = arr[end], arr[start]
    start += 1
    end -= 1

def rotate_in_place(arr, k):
  n = len(arr)
  k = k % n  # In case k is greater than array length

  #Step 1: Reverse the entire array
  reverse(arr, 0, n - 1)

  #Step 2: Reverse the first k elements
  reverse(arr, 0, k - 1)

  #Step 3: Reverse the remaining n - k elements
  reverse(arr, k, n-1)
```

```python
## Applying the function

arr = [1, 2, 3, 4, 5]
rotate_in_place(arr, 2)
print(arr)
```

## Sorting Algorithm

```python
# Bubble Sort

def bubble_sort(arr):
  n = len(arr)
  for i in range(n):
    #Last i elements are already sorted
    for j in range(0, n - i - 1):
      #Swap if the element found is greater than the next element
      if arr[j] > arr[j + 1]:
        arr[j], arr[j + 1] =  arr[j + 1], arr[j]
  return arr

#Example usage:
arr = [64, 34, 25, 12, 22, 11, 90]
print(bubble_sort(arr))
```

```python
##Let's rewrite bubble_sort

#First thing, define the function
def bubble_sort(arr):
  n = len(arr)

  for i in range(n):
    for j in range(0, n - i - 1):
      if arr[j] > arr[j + 1]:
        arr[j], arr[j + 1]= arr[j + 1], arr[j]
  return arr


#Usage
arr2 = [14, 10, 3, 20, 25, 3, 4, 60]
print(bubble_sort(arr2))
```

Buddle sort is not efficient for large data but space complexity is constant

Time complexity = O(n^2)
Space complexity = O(1)

```python
## Using Quick Sort, I will learn both Bubble and Quick Sort for the purpose of this interview

def quick_sort(arr):
  if len(arr) <= 1:
    return arr
  else:
    pivot = arr[len(arr) // 2]  #Choose the middle element as pivot
    left = [x for x in arr if x < pivot]  ## Elements less than pivot
    middle = [x for x in arr if x == pivot] ## Elements equal to pivot
    right = [x for x in arr if x > pivot] ## Elements greater than pivot
    return quick_sort(left) + middle + quick_sort(right)


# Example usage
arr = [10, 7, 8, 9, 1, 5]
print(quick_sort(arr))
```
